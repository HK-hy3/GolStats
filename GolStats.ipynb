```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("ASUSE_2021.csv", low_memory=False)

# Convert columns to numeric safely
for col in ['total_inventories', 'avg_employees_computers', 'avg_employees_inter']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Drop rows where key identifiers are missing
df = df.dropna(subset=['sector', 'nss_region'])

if df.empty:
    print("Dataset is empty after preprocessing.")
else:
    # 1. Sector-wise accident distribution
    plt.figure(figsize=(12, 6))
    sns.countplot(data=df, x='sector', order=df['sector'].value_counts().index, palette='viridis')
    plt.xticks(rotation=45)
    plt.title("Sector-wise Accident Distribution")
    plt.xlabel("Employment Sector")
    plt.ylabel("Number of Cases")
    plt.show()

    # 2. Geographic Distribution of Accidents (Heatmap)
    state_accidents = df.groupby('nss_region').size().reset_index(name='count')
    if not state_accidents.empty:
        plt.figure(figsize=(10, 6))
        sns.heatmap(state_accidents.pivot_table(index='nss_region', values='count'), cmap="coolwarm", annot=True)
        plt.title("Accident Distribution by NSS Region")
        plt.xlabel("NSS Region")
        plt.ylabel("Accident Count")
        plt.show()
    else:
        print("No data available for heatmap.")

    # 3. Financial Loss Due to Accidents
    plt.figure(figsize=(10, 5))
    df.groupby("sector")['total_inventories'].sum().sort_values().plot(kind="bar", colormap='plasma')
    plt.title("Total Financial Loss Due to Accidents by Sector")
    plt.xlabel("Sector")
    plt.ylabel("Total Financial Loss (in currency units)")
    plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (Replace with actual dataset)
df = pd.read_csv("ASUSE_2021.csv", low_memory=False)  # Ensure correct file path

# 1. Employment Sector Risk: Bar Chart for Road Accident-Prone Sectors
plt.figure(figsize=(12, 6))
sector_risk = df["sector"].value_counts().head(10)  # Top 10 sectors by accident records
sns.barplot(x=sector_risk.index, y=sector_risk.values, palette="Reds")
plt.xticks(rotation=45)
plt.xlabel("Employment Sector")
plt.ylabel("Number of Reported Cases")
plt.title("Top 10 Employment Sectors Involved in Road Accidents")
plt.show()

# 2. Geographic Distribution: Heatmap of Road Accidents by State
plt.figure(figsize=(12, 6))
state_accidents = df["nss_region"].value_counts().reset_index()
state_accidents.columns = ["State", "Reported Cases"]
state_accidents = state_accidents.pivot_table(values="Reported Cases", index="State")

sns.heatmap(state_accidents, cmap="Reds", annot=True, linewidths=0.5)
plt.title("State-Wise Road Accidents Heatmap")
plt.show()

# 3. Compensation & Legal Protection: Pie Chart of Compensation Status
if "compensation_status" in df.columns:
    compensation_counts = df["compensation_status"].value_counts()
    plt.figure(figsize=(8, 6))
    plt.pie(compensation_counts, labels=compensation_counts.index, autopct='%1.1f%%', colors=['red', 'green', 'blue'])
    plt.title("Compensation Status for Road Accident Victims")
    plt.show()

# 4. Infrastructure Issues: Road Safety Concerns Over the Years
if "year" not in df.columns:
    df["year"] = pd.to_datetime(df["sub_round"], errors='coerce').dt.year  # If a date column exists, adjust accordingly

yearly_trends = df["year"].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_trends.index, y=yearly_trends.values, marker='o', color='blue')
plt.xlabel("Year")
plt.ylabel("Reported Cases")
plt.title("Road Safety Concerns Over the Years")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from reportlab.lib.pagesizes import letter, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle
from reportlab.lib import colors

def csv_to_pdf(csv_file, pdf_file):
    # Read CSV file
    df = pd.read_csv(csv_file)

    # Convert DataFrame to list of lists
    data = [df.columns.tolist()] + df.values.tolist()

    # Create a PDF document
    pdf = SimpleDocTemplate(pdf_file, pagesize=landscape(letter))

    # Create a table
    table = Table(data)

    # Add style to table
    style = TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
        ('GRID', (0, 0), (-1, -1), 1, colors.black)
    ])

    table.setStyle(style)

    # Build the PDF
    pdf.build([table])

    print(f'PDF saved as {pdf_file}')

# Example usage
csv_to_pdf("example.csv", "output.pdf")
```
